## Step One: Collect Data 

We will be using data from the Sloan Digital Sky Survey (SDSS). There are some useful tutorials available online regarding how to query data for your analysis, 

* https://www.sdss4.org/dr17/tutorials/getimage/

Here is a column breakdown, 

* `objID` : Unique identifier for the astronomical object in the SDSS photometric catalog. 
* `ra` : Right Ascension: sky longitude.

```python
#!pip install astroquery astropy
import pandas as pd
from astroquery.sdss import SDSS
```

```python
query = """
SELECT TOP 300
    p.objID,
    p.ra,
    p.dec,
    p.u,
    p.g,
    p.r,
    p.i,
    p.z,
    s.class,
    s.z AS redshift
FROM PhotoObj AS p
JOIN SpecObj AS s
    ON p.objID = s.bestObjID
WHERE
    s.class IN ('STAR', 'GALAXY', 'QSO')
    AND p.u > 0
    AND p.g > 0
    AND p.r > 0
    AND p.i > 0
    AND p.z > 0
"""

results = SDSS.query_sql(query)

results
```
